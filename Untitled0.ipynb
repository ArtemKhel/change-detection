```python

N_RUNS = 10
eta = .5
lambda_ = .5
rulsif_alpha = 0.1

window_sizes = [(i, i, i // 2) for i in (10, 20, 30)]
sigmas = [i for i in range(5, 11)]
# thresholds = [2 ** i for i in range(-3, 4)]
thresholds = [.1, .2, .3, .4, .5, .75, 1, 2, 4, 8]
```

```python
def err_callback(err):
    print(err)
```

```python
def new_plot(ds, mu, t_list, s_history, change_points, save=False, file='/tmp/file.png'):
    fig, ax = plt.subplots(2, 1, figsize=(10, 5))

    ax[0].plot(ds)
    ax[0].vlines(change_points, np.nanmin(ds), np.nanmax(ds), color='red')

    ax[1].plot(t_list, s_history)
    ax[1].plot(t_list, [mu] * len(t_list), linestyle='dashed', color='m')
    ax[1].plot(t_list, [-mu] * len(t_list), linestyle='dashed', color='m')
    ax[1].set_xlim(ax[0].get_xlim())
    ax[1].vlines(change_points, np.nanmin(s_history), np.nanmax(s_history), color='r')
    if save:
        plt.savefig(file)
    else:
        plt.show()
```

```python
dsize = (100, 1)


def generate_data():
    return {  # TODO: size change / mean change for same ds?
        'normal': [
            normal(loc=0, scale=1, size=dsize),
            normal(loc=0, scale=5, size=dsize),
            # normal(loc=5, scale=1, size=dsize),
        ],
        'uniform': [
            uniform(low=-1, high=1, size=dsize),
            uniform(low=-5, high=5, size=dsize),
        ],
        'poisson': [
            poisson(lam=1, size=dsize),
            poisson(lam=5, size=dsize),
        ],
        'exp': [
            exponential(scale=5, size=dsize),
            exponential(scale=10, size=dsize),
        ],
        # 'cauchy': [
        #     standard_cauchy(size=dsize),
        #     standard_cauchy(size=dsize) * 5,
        # ],
    }
```

```python
def make_datasets(data, large=True, half=False):
    for i, fst in enumerate(data.keys()):
        for j, snd in enumerate(data.keys()):
            if half and j < i: continue
            intervals = (
                [
                    data[fst][0],
                    data[fst][1] if fst == snd else data[snd][0],
                    data[fst][0],
                ] + ([
                         data[fst][1] if fst == snd else data[snd][0],
                         data[fst][0]
                     ] if large else []))
            yield fst, snd, np.concatenate(intervals)
```

```python
def bruteforce_threshold(ds, n_ref, n_test, k, n_cp=2):
    kliep_thresholds = {}
    rulsif_thresholds = {}
    for threshold in sorted(thresholds):
        kl = KLIEP(mu=threshold, eta=eta, lambda_=lambda_, sigmas=sigmas)
        kl.run(ds, n_ref=n_ref, n_test=n_test, k=k)
        kliep_thresholds[threshold] = len(kl.change_points)
        if len(kl.change_points) == 0: break

    for threshold in sorted(thresholds):
        ru = RULSIF(mu=threshold, lambda_=lambda_, alpha=rulsif_alpha)
        ru.run(ds, n_ref=n_ref, n_test=n_test, k=k)
        rulsif_thresholds[threshold] = len(ru.change_points)
        if len(ru.change_points) == 0: break

    # print(kliep_thresholds, rulsif_thresholds)
    kliep_thresholds = [(tr, n) for tr, n in sorted(kliep_thresholds.items(), key=lambda x: abs(x[1] - n_cp))]
    rulsif_thresholds = [(tr, n) for tr, n in sorted(rulsif_thresholds.items(), key=lambda x: abs(x[1] - n_cp))]

    kl = kliep_thresholds[0]
    ru = rulsif_thresholds[0]

    kl = np.average(list(y[0] for y in filter(
        lambda x: (x[0] == kl[0] or x[1] != 0) and abs(x[1] - n_cp) == abs(kl[1] - n_cp),
        kliep_thresholds)))
    ru = np.average(list(y[0] for y in filter(
        lambda x: (x[0] == ru[0] or x[1] != 0) and abs(x[1] - n_cp) == abs(ru[1] - n_cp),
        rulsif_thresholds)))

    return kl, ru
```

```python
_data = generate_data()
pretrain_datasets = make_datasets(_data, half=True)
optimal_thresholds = {}


def _callback(res, d1, d2, ws):
    optimal_thresholds[frozenset((d1, d2, ws))] = res


with mp.Pool() as pool:
    for dist_1, dist_2, ds in pretrain_datasets:
        for ws in window_sizes:
            pool.apply_async(
                bruteforce_threshold,
                args=(ds, *ws),
                callback=partial(_callback, d1=dist_1, d2=dist_2, ws=ws),
                error_callback=err_callback
            )
            # break
        # break
    pool.close()
    pool.join()

with open('_opt_thresholds.pickle', 'wb') as f:
    pickle.dump(optimal_thresholds, file=f)
```

```python
# because multiprocessing rely on pickle and for some reason algorithm classes cannot be pickled
def run_kliep(*args):
    start = time()
    kl = KLIEP(**(args[0]))
    kl.run(*args[1:])
    end = time()
    return kl.t_list, kl.s_history, kl.change_points, end - start


def run_rulsif(*args):
    start = time()
    ru = RULSIF(**(args[0]))
    ru.run(*args[1:])
    end = time()
    return ru.t_list, ru.s_history, ru.change_points, end - start
```

```python
# results = []

def add_result(res, alg, dist, ws, dataset, threshold):
    new_plot(dataset, threshold, *res)
    # results.append((alg, dist, ws, dataset, threshold, *res))


with mp.Pool() as pool:
    for _ in range(N_RUNS):
        data = generate_data()
        datasets = make_datasets(data)

        for dist_1, dist_2, dataset in datasets:
            for window_size in window_sizes:
                kliep_threshold, rulsif_threshold = optimal_thresholds[frozenset((dist_1, dist_2, window_size))]

                callback = partial(add_result, dist=f'{dist_1}-{dist_2}', ws=window_size, dataset=dataset)

                kwargs = {'mu': kliep_threshold, 'eta': eta, 'lambda_': lambda_, 'sigmas': sigmas}
                pool.apply_async(
                    run_kliep,
                    args=(kwargs, dataset, *window_size),
                    callback=partial(callback, alg='kliep', threshold=kliep_threshold),
                    error_callback=err_callback
                )
                kwargs = {'mu': rulsif_threshold, 'lambda_': lambda_, 'alpha': rulsif_alpha}
                pool.apply_async(
                    run_rulsif,
                    args=(kwargs, dataset, *window_size),
                    callback=partial(callback, alg='rulsif', threshold=rulsif_threshold),
                    error_callback=err_callback
                )
                # break
            # break
        # break

    pool.close()
    pool.join()
```

```python
def calc_tp_fp(expected: list[int], actual: list[int], ws):
    r = sum(ws)
    # r = ws[1] + ws[2]
    tp = sum(1 for e in expected if any(e <= a <= (e + r) for a in actual))
    fp = sum(1 for a in actual if not any((a - r) < e < a for e in expected))
    fn = sum(1 for e in expected if not any(e < a < (e + r) for a in actual))
    return tp, fp, fn


def stats(df):
    tp = fp = fn = 0
    exp = [100 * i for i in range(1, 5)]
    n_cp = 4

    for i, r in df.iterrows():
        _tp, _fp, _fn = calc_tp_fp(expected=exp, actual=r.change_points, ws=r.ws)
        tp += _tp
        fp += _fp
        fn += _fn
    return (
        tp,
        fp,
        fn,
        tp / (tp + fn) if tp + fn else 0,
        tp / (tp + fp) if tp + fp else 0,
        # fp / (tp + fp) if tp + fp else 0,
        tp / (df.shape[0] * n_cp),
        # fp / (df.shape[0] * n_cp)
    )


alg_ws_df = pd.DataFrame(columns=('Alg', 'WS', 'Time', 'TP', 'FP', 'FN', 'Sensitivity', 'Precision', 'Accuracy'))
for alg in df.alg.unique():
    # print(f'{alg}:')
    for ws in df.ws.unique():
        s = stats(df.loc[(df.alg == alg) & (df.ws == ws)][['change_points', 'ws']])
        t = df.loc[(df.alg == alg) & (df.ws == ws)][['time']].squeeze().mean()
        alg_ws_df.loc[len(alg_ws_df)] = (alg, ws, t, *s)
alg_ws_df
```
